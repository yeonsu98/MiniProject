```python
# 파일명에 '_'이 붙어있거나, 순서가 맞지 않아 순서대로 파일명 rename
for l in labels:
    path = TRAIN_PATH + '/' + l
    name = os.listdir(path)
    i = 0
    for n in name:
        src = os.path.join(path, n)
        new = os.path.join(path, str(i) + '_.jpg')
        os.rename(src, new)
        i+=1
        
for l in labels:
    path = TRAIN_PATH + '/' + l
    name = os.listdir(path)
    i = 0
    for n in name:
        src = os.path.join(path, n)
        dst = str(i) + '.jpg'
        dst = os.path.join(path, dst)
        os.rename(src, dst)
        i+=1
```
