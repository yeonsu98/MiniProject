```python
# pycaret 라이브러리 import
from pycaret.regression import *
```

```python
# setup으로 init
clf = setup(data=df, target='ET', train_size=0.8)
```

```python
# ML 모델 성능 비교
comp = compare_models(sort = 'RMSE', n_select=5)
```

```python
# 하이파라미터 튜닝
tuned_top5 = [tune_model(i) for i in comp]
```

```python
# blending 모델 생성
blender_top5 = blend_models(estimator_list=tuned_top5)
```

```python
# 예측 성능 확인
final_model = finalize_model(blender_top5)
prediction = predict_model(final_model, data=df_test)
```
