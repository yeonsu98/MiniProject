```python
# 데이터 불러오기
from pycaret.datasets import get_data

juice = get_data('juice')
```

```python
# set init
from pycaret.classification import *


clf = setup(data=juice, target='Purchase')
```

```python
# 정확도 상위 5위 모델 비교
top5 = compare_models(sort='Accuracy', n_select=5)
```

```python
# 하이퍼파라미터 튜닝
tuned_top5 = [tune_model(i) for i in top5]
```

```python
# 튜닝된 상위 5개의 모델 블렌딩
blender_top5 = blend_models(estimator_list=tuned_top5)
```

```python
final_model = finalize_model(blender_top5)
prediction = predict_model(final_model, data=juice.iloc[-100:])
```

```python
from pycaret.utils import check_metric

check_metric(prediction['Purchase'], prediction['Label'], metric = 'Accuracy') # 0.81
```
