### 가. LinearRegression 부터 시작해봅시다.

#### [실습문제1] 머신러닝_1
* Train과 Test로 나눈 데이터를 기준으로 LinearRegression 모델링을 진행하고 평가를 해주세요.
* 성능지표 : RMSE, R-squared Score

```python
# 아래에 실습코드를 작성하세요.
def pred_result(true,pred):
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import r2_score
    print('RMSE : %.5f' % mean_squared_error(true, pred)**0.5)
    print('R2-Score : %.5f' % r2_score(true, pred))
```

```python
# 아래에 실습코드를 작성하세요.
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_x, train_y)
pred = lr.predict(test_x)
pred_result(test_y, pred)
```

## 이번엔 딥러닝 모델링을 해봅시다.

#### [실습문제9] 딥러닝

* tensorflow 라이브러리를 활용한 모델을 자유롭게 만들어보세요.

```python
!pip install tensorflow 
```

```python
# 아래에 실습코드를 작성하세요.
from tensorflow import keras
import tensorflow as tf
```

```python
# 아래에 실습코드를 작성하세요.
#1. 세션 클리어
keras.backend.clear_session()

#2. 레이어 엮기
il = keras.layers.Input(shape=(train_x.shape[1]))
hl = keras.layers.Dense(128, activation='relu')(il)
hl = keras.layers.Dense(128, activation='relu')(hl)
hl = keras.layers.Dense(64, activation='relu')(hl)
ol = keras.layers.Dense(1)(hl)

# 3. 모델 생성
model = keras.models.Model(il, ol)

# 4. 모델 컴파일
model.compile(loss='mse', optimizer='adam')

# 5. 모델 구조
model.summary()
```

```python
# 아래에 실습코드를 작성하세요.
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# 아래에 실습코드를 작성하세요.
es = EarlyStopping(monitor='val_loss',
                  patience=5,
                  min_delta=0,
                  verbose=1)

history = model.fit(train_x, train_y, epochs=1000, callbacks=[es], validation_split=0.2, verbose=1)
```

```python
# 아래에 실습코드를 작성하세요.
pred = model.predict(test_x)
pred_result(test_y, pred)
```
