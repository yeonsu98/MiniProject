# 1. 데이터 전처리

#### [실습문제1] 열 추가 
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다. 

```python
# 1416 강서구 공항동
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'])
df_test['기준일ID'] = pd.to_datetime(df_test['기준일ID'])
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total['year'] = df_total['기준일ID'].dt.year
df_total['month'] = df_total['기준일ID'].dt.month
df_total['day'] = df_total['기준일ID'].dt.day
df_test['year'] = df_test['기준일ID'].dt.year
df_test['month'] = df_test['기준일ID'].dt.month
df_test['day'] = df_test['기준일ID'].dt.day
```

```python
df_total['dayofweek'] = df_total['기준일ID'].dt.dayofweek
df_test['dayofweek'] = df_test['기준일ID'].dt.dayofweek
```

```python
df_total
```

#### [실습문제2] 불필요한 열 제거 및 열 이름 변경
* df_total과 df_test에서 '총생활인구수'외에 불필요한 열을 제거 하도록 하겠습니다.
* 시간 데이터를 남기고, '시간대구분' column 을 'hour'로 rename 하도록 하겠습니다. 

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
del_col = '기준일ID'

df_total.drop(del_col, axis=1, inplace=True)
df_test.drop(del_col, axis=1, inplace=True)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total.rename(columns={'시간대구분' : 'hour'}, inplace=True)
df_test.rename(columns={'시간대구분' : 'hour'}, inplace=True)
```

```python
men = []
women = []
for i in range(2, 16):
    print('남', df_total.iloc[:, i].mean())
    men.append(df_total.iloc[:, i].mean())
    print('여', df_total.iloc[:, i+14].mean())
    women.append(df_total.iloc[:, i+14].mean())
    
age_col = df_total.columns[2:16]
age_name = []
for a in age_col:
    age_name.append(a[2:])
```

```python
import matplotlib.pyplot as plt
plt.title("나이대별 남자 평균 생활인구수")
plt.plot(age_name, men, marker='o')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title("나이대별 여자 평균 생활인구수")
plt.plot(age_name, women, marker='o')
plt.xticks(rotation=90)
plt.show()
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

```python
df_total.loc[(df_total['month']==10) & (df_total['year']==2019)]['day'].unique()
```

```python
# 작년 데이터 가져와서 넣기
df_2018_10 = df_total.loc[(df_total['month']==10) & (df_total['year']==2018) & (df_total['day'].isin(range(15, 28)))]
df_2018_10['year'] = 2019
df_2018_10
```

```python
df_train = pd.concat([df_total, df_2018_10], axis=0)
df_train
```

```python
df_train.loc[(df_train['month']==10) & (df_train['year']==2019)]['day'].unique()
```

```python
df_train.sort_values(by=['year', 'month', 'day'], inplace=True)
```

```python
df_train.loc[(df_train['month']==10) & (df_train['year']==2019)]['day'].unique()
```

```python
df_train['weekend'] = 0
df_test['weekend'] = 0
```

```python
df_train.loc[df_train['dayofweek'].isin([5, 6])]['weekend'] = 1
df_test.loc[df_test['dayofweek'].isin([5, 6])]['weekend'] = 1
```

```python
df_final = df_train.iloc[:, [0, 1, -2]]
```

```python
df_final
```

```python
df_test = df_test.iloc[:, [0, 1, -2]]
df_test
```

```python
df_final['하루전총생활인구수'] = df_final['총생활인구수'].shift(24)
df_final['일주일전총생활인구수'] = df_final['총생활인구수'].shift(24*7)
df_final['한달전총생활인구수'] = df_final['총생활인구수'].shift(24*30)
#df_final['두달전총생활인구수'] = df_final['총생활인구수'].shift(1416)

df_test['하루전총생활인구수'] = df_test['총생활인구수'].shift(24)
df_test['일주일전총생활인구수'] = df_test['총생활인구수'].shift(24*7)
df_test['한달전총생활인구수'] = df_test['총생활인구수'].shift(24*30)
#df_test['두달전총생활인구수'] = df_test['총생활인구수'].shift(1416)
```

```python
df_test
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_final['24시간동안의평균인구수'] = df_final['총생활인구수'].rolling(24).mean()
df_final['일주일동안의평균인구수'] = df_final['총생활인구수'].rolling(24*7).mean()
df_final['한달동안의평균인구수'] = df_final['총생활인구수'].rolling(24*30).mean()
#df_final['두달동안의평균인구수'] = df_final['총생활인구수'].rolling(1416).mean()
df_test['24시간동안의평균인구수'] = df_test['총생활인구수'].rolling(24).mean()
df_test['일주일동안의평균인구수'] = df_test['총생활인구수'].rolling(24*7).mean()
df_test['한달동안의평균인구수'] = df_test['총생활인구수'].rolling(24*30).mean()
#df_test['두달동안의평균인구수'] = df_test['총생활인구수'].rolling(1416).mean()
```
