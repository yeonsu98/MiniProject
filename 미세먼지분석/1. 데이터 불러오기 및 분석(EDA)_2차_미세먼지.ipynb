```python
air_21['측정일시'] = air_21['측정일시'].astype('int')
t = air_21['측정일시']
t
```

```python
temp = []
for i in t:
    if int(i%100) == 24:
        i += 76
    temp.append(i)
```

```python
for i in range(len(temp)):
    if temp[i] == 2021013200:
        temp[i] = 2021020100
    elif temp[i] == 2021022900:
        temp[i] = 2021030100
    elif temp[i] == 2021033200:
        temp[i] = 2021040100
    elif temp[i] == 2021043100:
        temp[i] = 2021050100
    elif temp[i] == 2021053200:
        temp[i] = 2021060100
    elif temp[i] == 2021063100:
        temp[i] = 2021070100
    elif temp[i] == 2021073200:
        temp[i] = 2021080100
    elif temp[i] == 2021083200:
        temp[i] = 2021090100
    elif temp[i] == 2021093100:
        temp[i] = 2021100100
    elif temp[i] == 2021103200:
        temp[i] = 2021110100
    elif temp[i] == 2021113100:
        temp[i] = 2021120100
    elif temp[i] == 2021123200:
        temp[i] = 2022010100
temp
```

```python
air_21['측정일시'] = temp
```

```python
# 측정일시를 datetime으로 변환
from datetime import *
format = '%Y%m%d%H'
air_21['측정일시'] = pd.to_datetime(air_21['측정일시'].astype('str'), format=format)
```
