# (실습준비) 데이터 불러오기부터

```python
import pandas as pd

# 학습 데이터 로딩
weather_21 = pd.read_csv("weather_2021.csv", sep = ',', encoding='cp949')
weather_22 = pd.read_csv("weather_2022.csv", sep = ',', encoding='cp949')
```

```python
air_21 = pd.read_csv("air_21_prep.csv", sep=',', index_col = 0, encoding = 'utf-8' )
air_22 = pd.read_csv("air_22_prep.csv", sep=',', index_col = 0, encoding = 'utf-8' )
```

```python
from datetime import *
from sklearn.impute import KNNImputer
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
weather_del_col = ['기온 QC플래그', '풍속 QC플래그', '풍향 QC플래그', '습도 QC플래그', 
                   '현지기압 QC플래그', '해면기압 QC플래그', '지면상태(지면상태코드)', 
                   '지면온도 QC플래그', '강수량 QC플래그', '일조 QC플래그', '3시간신적설(cm)',
                   '일조 QC플래그', '일사 QC플래그', '운형(운형약어)', '현상번호(국내식)']
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# 결측치가 행의 수만큼 있는 열은 제거
weather_21 = weather_21.drop(weather_del_col, axis=1)
weather_22 = weather_22.drop(weather_del_col, axis=1)
```

```python
# 적설량은 0으로 대체
weather_21['적설(cm)'] = weather_21['적설(cm)'].fillna(0)
```

```python
# 결측치 처리
col = weather_21.columns[3:]

imputer = KNNImputer(n_neighbors=3)
filled1 = imputer.fit_transform(weather_21[col])
weather_21[col] = filled1
weather_21.isna().sum()
```

```python
# 결측치 처리
col = weather_22.columns[3:]

imputer = KNNImputer(n_neighbors=3)
filled1 = imputer.fit_transform(weather_22[col])
weather_22[col] = filled1
weather_22.isna().sum()
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
weather_21['time'] = pd.to_datetime(weather_21['일시'].astype('str'))
weather_22['time'] = pd.to_datetime(weather_22['일시'].astype('str'))
```

```python
air_21['time'] = pd.to_datetime(air_21['time'].astype('str'))
air_22['time'] = pd.to_datetime(air_22['time'].astype('str'))
```

```python
df_21 = pd.merge(air_21, weather_21, on ='time')
df_22 = pd.merge(air_22, weather_22, on='time')
```

```python
df_21
```

```python
del_col = ['측정소명', '측정일시', '지점', '지점명', '일시']

df_21 = df_21.drop(del_col, axis=1)
df_22 = df_22.drop(del_col, axis=1)
```

```python
df_21['aft1h_PM10'] = df_21['PM10'].shift(-1, fill_value= 20.0)
df_22['aft1h_PM10'] = df_22['PM10'].shift(-1, fill_value=11.0)
df_22
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다
col = ['time', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25', '기온(°C)', '강수량(mm)',
       '풍속(m/s)', '풍향(16방위)', '습도(%)', '증기압(hPa)', '이슬점온도(°C)', '현지기압(hPa)',
       '해면기압(hPa)', '일조(hr)', '일사(MJ/m2)', '적설(cm)', '전운량(10분위)',
       '중하층운량(10분위)', '최저운고(100m )', '시정(10m)', '지면온도(°C)',
       '5cm 지중온도(°C)', '10cm 지중온도(°C)', '20cm 지중온도(°C)', '30cm 지중온도(°C)', 'aft1h_PM10']

df_21 = df_21[col]
df_22 = df_22[col]
```

```python
target = 'aft1h_PM10'
train_x = df_21.drop(target, axis=1)
train_y = df_21[target]
test_x = df_22.drop(target, axis=1)
test_y = df_22[target]
```

```python
train_x.to_csv('train_x2.csv', encoding='UTF-8', index=False)
train_y.to_csv('train_y2.csv', encoding='UTF-8', index=False)
test_x.to_csv('test_x2.csv', encoding='UTF-8', index=False)
test_y.to_csv('test_y2.csv', encoding='UTF-8', index=False)
```
