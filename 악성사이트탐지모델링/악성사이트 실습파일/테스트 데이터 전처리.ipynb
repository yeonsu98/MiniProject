```python
df = pd.read_csv('test_dataset_v01.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# 데이터가 모두 0인 칼럼('url_chinese_present', "html_num_tags('applet')") 제거
del_col = ['url_chinese_present', "html_num_tags('applet')"]
df = df.drop(del_col, axis=1)
```

```python
df.isna().sum()
```

```python
df['range_url'] = (df['url_len'] / 10).astype('int')
```

```python
idx = df.loc[df['url_path_len'].isna()].index
```

```python
na = df.loc[df['url_path_len'].isna()]
```

```python
na.isna().sum()
```

```python
na
```

```python
t = df.groupby('range_url')[['url_path_len']].mean()
```

```python
a = pd.merge(na, t, on='range_url')
```

```python
a.isna().sum()
```

```python
a = a.sort_values('range_url')
a['url_path_len_y'] = a['url_path_len_y'].fillna(method='ffill')
```

```python
a.isna().sum()
```

```python
a = a.drop('url_path_len_x', axis=1)
```

```python
a = a.rename(columns={'url_path_len_y' : 'url_path_len'})
```

```python
a.shape
```

```python
df = df.drop(idx, axis=0)
```

```python
df = pd.concat([df, a])
```

```python
df.shape
```
